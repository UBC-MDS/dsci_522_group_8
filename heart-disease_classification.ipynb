```python
#Import Packages
import altair as alt
import pandas as pd
import numpy as np
import statistics
from sklearn.impute import KNNImputer

from ucimlrepo import fetch_ucirepo 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
## Load Dataset
```

```python
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features
y = heart_disease.data.targets
  
# metadata 
## print(heart_disease.metadata) 
  
# variable information 
var = pd.DataFrame(heart_disease.variables) 
display(var)

#column names
colnames = pd.Series(X.columns)
```

```python
#Imputing Missing Variables using kNN Imputer
imputer = KNNImputer(n_neighbors = 4)
df = pd.DataFrame(imputer.fit_transform(X), columns = colnames)
df["target"] = y
```

<h3> <center> Exploratory Data Analysis </center> </h3>

```python
#Summary Statistics
display(pd.DataFrame(df.describe()).T)
```

```python
#Distribution of Values
##Note: drop some uncessary features

alt.Chart(df).mark_bar(size = 12, color = "teal").encode(
x = alt.X(alt.repeat()).bin(maxbins=5).scale(zero=False),
y = alt.Y('count()').scale(zero=False).title("Count")
).properties(
    width = 100, height = 100
).repeat(
    list(df.columns)
)
```

```python
#Pairwise Plot
##Note: drop some uncessary features
alt.Chart(df).mark_square(size = 12).encode(
    alt.X(alt.repeat("row")).scale(zero=False).type("quantitative"),
    alt.Y(alt.repeat("column")).scale(zero=False).type("quantitative"),
    color = alt.Color("target").scale(scheme='viridis')
).properties(height = 100, width = 100).repeat(
    row = list(df.columns.drop("target")),
    column = list(df.columns.drop("target"))
)
```
